## Residential Smart Meter Data Analysis

This notebook presents a brief analysis on a set of anonymous residential smart meter readings from several hundred meters over a 17-day period, provided by Octopus Energy, at a 30-minute resolution. We will be looking to:
* Quantify various interpolation approaches to handle missing readings
* Explore several approaches to visualising collections of smart meter data 
* Apply some simple dimensionality reduction to improve ease of computation
* Group each household using k-means clustering, and identify any interesting usage trends among and between these groups

Using the Future Energy Associates Taiffscanner API, we will also investigate:
* The potential savings for each household should they move to the Octopus Agile tariff from the cheapest fixed and variable tariff respectively
* Identifying which clusters of consumers would benefit the most from moving to the Agile tariff

```python
def aggregate_raw_data():
    """ 
    Load all CSV files in the raw data directory
    and concatenate into a single DataFrame instance.
    """
    
    agg_files_list = list()
    
    for raw_file_path in raw_meter_files:
        raw_meter_data = pd.read_csv(Path(raw_data_path.joinpath(raw_file_path)))
        agg_files_list.append(raw_meter_data)
        
    return pd.concat(agg_files_list)

meter_readings = aggregate_raw_data()

print(f"Number of unique meters: {meter_readings.meter_identifier.nunique()}")
```

At this stage, we also want to check the consistency of each meter's readings through the measurement period and accommodate missing values where possible.

```python
def calculate_num_missing_periods(df, period = PERIOD):
    """
    For a given Series of datetime values, calculate the number of missing periods
    by subtracting the number of observations from the number of expected periods 
    given the min and max datetimes of the Series.
    """

    expected_periods = 1 + (df.max() - df.min()) // period

    return expected_periods - len(df)


def calculate_num_missing_periods_per_meter(meter_readings, period = PERIOD):
    """
    Returns an ordered DataFrame of meter IDs against the 
    number of missing meter readings periods.
    """

    missing_periods_dict = dict()

    for meter_id in meter_readings.meter_identifier.unique():

        num_missing_periods = calculate_num_missing_periods(meter_readings[meter_readings.meter_identifier == meter_id].interval_start)
        missing_periods_dict[meter_id] = num_missing_periods

    return (pd.DataFrame.from_dict(missing_periods_dict, 
                                   orient = "index",
                                   columns = ["num_missing_periods"])
            .sort_values("num_missing_periods", ascending = False))

missing_periods_per_meter = calculate_num_missing_periods_per_meter(meter_readings)
```

The peaks in missing periods in the first three multiples of 48 suggest that when meters are missing any periods, it is likely one (or more) entire days at a time.

We can expand on this by looking at the distribution of the size of the missing periods:

```python
def calculate_missing_period_distribution(meter_readings):
    """
    Returns a dict of timedeltas indicating the distribution 
    of time deltas between each meter readings, for each unique meter.
    """

    meter_ids = meter_readings.meter_identifier.unique()
    
    missing_period_dict = {meter_id: [] for meter_id in meter_ids}
    
    for meter_id in meter_ids:
        meter_periods = meter_readings[meter_readings.meter_identifier == meter_id].interval_start
        missing_period_dict[meter_id] = meter_periods.sort_values().diff(1).dropna()

    return missing_period_dict

missing_period_dict = calculate_missing_period_distribution(meter_readings)
```

```python
# we're not interested in the half-hour time deltas, so we can remove those

missing_period_values = list(chain.from_iterable(missing_period_dict.values()))
missing_period_values = pd.Series([period // PERIOD for period in missing_period_values if period != PERIOD])
```

Almost all missing periods occur in blocks of one day, although there are enough intra-day gaps to warrant considering an effective interpolation or drop-out strategy

```python
def calculate_missing_period_starts(meter_readings):
    """
    Return a sorted DataFrame of datetimes and a counter 
    indicating the number of missing periods starting from 
    that datetime.
    """

    meter_ids = meter_readings.meter_identifier.unique()
    
    missing_period_start_list = list()

    for meter_id in meter_ids:
        
        meter_periods = (meter_readings[meter_readings.meter_identifier == meter_id]
                         .sort_values("interval_start")
                         .reset_index()
                         .interval_start)
        
        meter_periods_diff = meter_periods.diff(1).dropna()
        idxs = meter_periods_diff[meter_periods_diff != PERIOD].index
        missing_period_start_list.extend(meter_periods.loc[idxs].values)

    counter_dict = {counter[0]: counter[1] for counter in list(Counter(missing_period_start_list).items())}
    counter_df = pd.DataFrame.from_dict(counter_dict, orient="index")
    counter_df.columns = ["num_missing_periods"]

    return counter_df

missing_periods_start_df = calculate_missing_period_starts(meter_readings)
```

The majority of missing periods start at midnight - corresponding to whole missing days. Smaller gaps are more irregularly distributed, so let's try interpolating from the smallest gaps upwards until an unacceptable prediction error is found.

```python
def test_interpolation_methods(day_reading, gap_length, gap_starting_period, plot = True):
    """
    Take a day's sample of complete data from the meter readings
    provided, remove a consecutive series of readings of length
    gap_length, and test several interpolation methods to infill
    the missing values.
    """
    
    # check that the provided day_reading is a complete day of measurements
    assert len(day_reading) == pd.Timedelta(1, "d") // PERIOD, "Please ensure a complete day of readings is passed."
    assert day_reading.interval_start.diff(1).dropna().unique()[0] == PERIOD, "Please ensure all measurement periods are consecutive."
    
    # extract a segment from the time series
    gap_ending_period = gap_starting_period + gap_length
    day_reading_with_missing = day_reading.copy().reset_index()
    day_reading_with_missing.loc[gap_starting_period:gap_ending_period-1, "kWh_per_half_hour"] = np.nan

    # set indexes to interval_start for simpler plotting
    day_reading = day_reading.set_index("interval_start")
    day_reading_with_missing = day_reading_with_missing.set_index("interval_start")

    linear_interp = day_reading_with_missing['kWh_per_half_hour'].interpolate()
    poly_interp = day_reading_with_missing['kWh_per_half_hour'].interpolate(method='polynomial', order=3)
    spline_interp = day_reading_with_missing['kWh_per_half_hour'].interpolate(method='spline', order=3)

    # we only want to record the MAEs for the gap period of missing values
    mae_results = {
        "linear": np.round(mean_absolute_error(linear_interp.iloc[gap_starting_period:gap_ending_period], 
                                               day_reading.iloc[gap_starting_period:gap_ending_period]['kWh_per_half_hour']), 
                           4),
        "poly": np.round(mean_absolute_error(poly_interp.iloc[gap_starting_period:gap_ending_period],
                                             day_reading.iloc[gap_starting_period:gap_ending_period]['kWh_per_half_hour']),
                         4),
        "spline": np.round(mean_absolute_error(spline_interp.iloc[gap_starting_period:gap_ending_period], 
                                               day_reading.iloc[gap_starting_period:gap_ending_period]['kWh_per_half_hour']), 
                           4),
        "naive_mean": np.round(mean_absolute_error((day_reading_with_missing["kWh_per_half_hour"]
                                                   .fillna(day_reading_with_missing["kWh_per_half_hour"]
                                                           .mean())
                                                    .iloc[gap_starting_period:gap_ending_period]),
                                                  day_reading.iloc[gap_starting_period:gap_ending_period]['kWh_per_half_hour']),
                               4)
    }

    if plot:
        fig, ax = plt.subplots(figsize=(12, 4))
        
        # plot different interpolations
        ax.plot(day_reading['kWh_per_half_hour'], label='True Reading')
        ax.plot(linear_interp.iloc[gap_starting_period - 1:gap_ending_period + 1], label='Linear')
        ax.plot(poly_interp.iloc[gap_starting_period - 1:gap_ending_period + 1], label='Polynomial')
        ax.plot(spline_interp.iloc[gap_starting_period - 1:gap_ending_period + 1], label='Spline')

        ax.set_title(f"Interpolations for Gap Length {gap_length} Starting at Period {gap_starting_period}")
        
        ax.axvspan(day_reading.index[gap_starting_period - 1], day_reading.index[gap_ending_period], color='red', alpha=0.1)
        ax.xaxis.set_major_locator(mdates.HourLocator())
        ax.xaxis.set_major_formatter(mdates.DateFormatter("%H"))
        ax.grid()
        ax.set_xlabel('Hour of Day')
        ax.set_ylabel('Energy (kWh)')
        ax.legend()

    return mae_results

for gap_starting_period, gap_length in product([12, 24, 36], [2, 3, 4]):
    test_interpolation_methods(meter_readings.iloc[:48].sort_values("interval_start").reset_index(), 
                               gap_length = gap_length, 
                               gap_starting_period = gap_starting_period,
                               plot = True)
```

As these smart meter readings are very non-linear and jagged, polynomial regression suffers the most, routinely under-/over-shooting the actual meter readings. We can quantify the MSE for each method a little more rigorously for a range of days, gap lengths, and start periods to settle on an approriate infilling strategy:

```python
def identify_complete_days(meter_readings):
    """
    Generate a list of meter-datetime pairs that indicate the start
    of a complete day of readings beginning at midnight.
    """ 

    complete_days_dict = dict()
    
    for meter_id in tqdm(meter_readings.meter_identifier.unique(), 
                         desc = "Iterating Meter IDs"):
        
        complete_days_dict[meter_id] = []
        
        for start_datetime in pd.date_range(start = meter_readings.date.min(),
                                            end = meter_readings.date.max(),
                                            freq = "D",
                                            tz = "utc"):
                                                
            meter_sample = meter_readings.loc[(meter_readings.meter_identifier == meter_id) & 
                                              (meter_readings.interval_start >= start_datetime) &
                                              (meter_readings.interval_end <= start_datetime + pd.Timedelta(1, "D"))]
                                                
            if meter_sample.shape[0] == pd.Timedelta(1, "D") // PERIOD:
                complete_days_dict[meter_id].append(start_datetime)

    return complete_days_dict


def return_all_complete_days(meter_readings, complete_days_dict = None):
    """
    Return a reduced DataFrame containing only days of complete readings for
    each meter.
    """

    if not complete_days_dict:
        complete_days_dict = identify_complete_days(meter_readings)

    complete_days_df = pd.DataFrame()
    
    for meter_id, start_datetimes in tqdm(complete_days_dict.items(),
                                          desc = "Aggregating Complete Days"):
        for start_datetime in start_datetimes:
            complete_days_df = pd.concat([complete_days_df,
                                          meter_readings.loc[(meter_readings.meter_identifier == meter_id) & 
                                                             (meter_readings.interval_start >= start_datetime) &
                                                             (meter_readings.interval_end <= start_datetime + pd.Timedelta(1, "D"))]])

    return complete_days_df
    

complete_days_df = return_all_complete_days(meter_readings)
```

```python
def sample_interpolate_performance(meter_readings, meter_samples: int = 100):
    """
    Collects a sample of MAE metrics for four interpolation methods over a meter_samples
    number of complete day readings. 
    """
    
    mae_results_dict = dict()
    
    # sample 100 complete days to sample each gap_length over
    meter_sample_ids = np.random.randint(low = 0, 
                                         high = 2500, 
                                         size = meter_samples) * 48
    
    for gap_length in range(2, 9):
    
        mae_results_dict[gap_length] = {"linear": [],
                                        "poly": [],
                                        "spline": [],
                                        "naive_mean": []}
        
        for meter_sample in tqdm(meter_sample_ids,
                                 desc = f"Iterating for gap length {gap_length}"):
            
            gap_starting_period = np.random.randint(6, 38)

            
            mae_results = test_interpolation_methods(day_reading = meter_readings.iloc[meter_sample:meter_sample+48].sort_values("interval_start"),
                                                     gap_length = gap_length,
                                                     gap_starting_period = gap_starting_period,
                                                     plot = False)
    
            for method, mae in mae_results.items():
                mae_results_dict[gap_length][method].append(mae)

    return mae_results_dict

mae_results_dict = sample_interpolate_performance(complete_days_df, meter_samples = 100)
```

```python
fig, ax = plt.subplots(figsize = (8, 6))

(pd.DataFrame.from_dict(mae_results_dict)
             .applymap(np.mean)
             .T
             .plot
             .bar(ax = ax))

ax.grid()
ax.set_xlabel('Gap Length')
ax.set_ylabel('Average MAE')
ax.set_title("Average MAE against Missing Measurement Length")
ax.legend(title = "Interpolation Method",
          labels = ["Linear", "Poly (Degree 3)", "Spline (Degree 3)", "Naive Mean"],
          bbox_to_anchor=(1, 1), 
          loc='upper left')
```

As expected, the MAE increases with the size of the missing measurement gap, although polynomial interpolation seems to suffer far more than the other methods as the gap length grows. 

In all cases, linear interpolation appears to outperform the other methods, despite the naive mean approach giving surpisingly good results against the more complex methods.

## Dimensionality Reduction

Visualise some PCA to see if any patterns emerge that may aid clustering further on. But first, let's see which kind of scaling may be most appropriate for the highly skewed and imbalances collection of meter readings we have:

```python
fig, ax = plt.subplots(figsize = (8, 6))

complete_days_df.kWh_per_half_hour.plot.hist(bins=25, 
                                             ax = ax,
                                             zorder = 3)
ax.legend("")
ax.grid(zorder = 0)
ax.set_xlabel("Consumption per Period (kWh)")
ax.set_ylabel("Number of Recorded Periods")
ax.set_title("Histogram of Recorded Consumption per Period")
```

Given the distribution is high log-normal and always positive (with a slight adjustment to 0 values), a Box-Cox transform may be most appropriate:



```python
complete_days_df["kWh_per_half_hour"] = complete_days_df.kWh_per_half_hour.replace(0.0, 1e-6)
```

```python
scaler = PowerTransformer(method = "box-cox")
scaled_readings = pd.DataFrame(scaler.fit_transform(complete_days_df.set_index("interval_start").kWh_per_half_hour.values.reshape(-1, 1)))

fig, ax = plt.subplots(figsize = (8, 6))

scaled_readings.plot.hist(bins=25, 
                          ax = ax,
                          zorder = 3)
ax.legend("")
ax.grid(zorder = 0)
ax.set_xlabel("Consumption per Period (kWh)")
ax.set_ylabel("Number of Recorded Periods")
ax.set_title("Histogram of Recorded Consumption per Period after Box-Cox Scaling")
```

```python
# # include the scaled consumtption readings as a new column in the DataFrame
complete_days_df["scaled_kWh"] = scaled_readings.values

# # to perform PCA, we want to pivot the dataframe such that each row represents a single day of readings
pca = PCA(n_components = 48, svd_solver="full")
scaled_readings_pca = pd.DataFrame(pca.fit_transform(complete_days_df.pivot(index = ["meter_identifier", "date"], columns = "time", values = "scaled_kWh")))

# plot the explained variance ratio per component
fig, ax = plt.subplots(figsize = (8, 6))

(pd.DataFrame(index = np.arange(1, 49), 
             data = pca.explained_variance_ratio_.cumsum(),
             columns = ["Cumulative Variance"])
    .plot(marker = "o",
         ax = ax))

ax.set_ylabel("Cumulative Explained Variance")
ax.set_xlabel("Number of PCA Components")
ax.legend("")
ax.grid()

# indicate 90% explained variance
ax.axhline(y = 0.9, 
           xmin = 0, 
           xmax = 50, 
           color = "r")
```

Approx. 18 components explain 90% of the variance, which may help with curse of dimensionality when clustering.

```python
def hinton(matrix, max_weight=None, ax=None):
    
    #Draw Hinton diagram for visualizing a weight matrix.
    ax = ax if ax is not None else plt.gca()

    if not max_weight:
        max_weight = 2**np.ceil(np.log(np.abs(matrix).max())/np.log(2))

    ax.set_aspect('equal', 'box')
    ax.xaxis.set_major_locator(plt.NullLocator())
    ax.yaxis.set_major_locator(plt.NullLocator())

    for (x, y), w in np.ndenumerate(matrix):
        color = 'red' if w > 0 else 'blue'
        size = np.sqrt(np.abs(w))
        rect = plt.Rectangle([x - size / 2, y - size / 2], size, size,
                             facecolor=color, edgecolor=color)
        ax.add_patch(rect)

    nticks = matrix.shape[0]
    ax.xaxis.tick_top()
    ax.set_xticks(range(48))
#    ax.set_xticklabels(list(matrix.columns))
    ax.set_yticks(range(nticks))
    ax.set_yticklabels(list(matrix.columns))
    ax.grid(False)

    ax.set_ylabel("PCA Component")
    ax.set_title("Original Feature")
    
    ax.autoscale_view()
    ax.invert_yaxis()
```

A Hinton diagram on the loadings (the product of the eigenvectors scaled by the sqrt of their corresponding eigenvalues) shows us the covariance between each PCA component and feature. As expected, the first component has the highest cumulative covariance with the features, decreasing to the 48th component as less and less variance is explained.

Of particular interest in the below is the wave-like nature of each component, with a lower frequency for the highest-variance components, gradually increasing until the frequency becomes so high that each feature alternates between positive and negative covariance as it progresses through each period of the day. This behaviour suggets that a Fourier transform might also give us some more information, but we will leave that to future work.

```python
components = pd.DataFrame(pca.components_)
loadings = pd.DataFrame(pca.components_).apply(lambda x: x * np.sqrt(pca.explained_variance_))

fig, ax = plt.subplots(figsize = (10, 10))
hinton(loadings.T, ax = ax)
```

## Clustering

With the data sufficiently scaled and reduced, we will now determine the optimum number of clusters for each day and meter using elbow and silhouette plots.

```python
# begin by clustering days pm 18 PCA components, independent of meter

pca_18 = PCA(n_components = 18, svd_solver="full")
scaled_readings_pca_18 = pd.DataFrame(pca_18.fit_transform(complete_days_df.pivot(index = ["meter_identifier", "date"], columns = "time", values = "scaled_kWh")))

cluster_range = range(2,16)
cluster_errors = []

for num_clusters in cluster_range:
    clusters = KMeans(n_clusters = num_clusters, n_init = 10).fit(scaled_readings_pca_18)
    cluster_errors.append(clusters.inertia_)

fig, ax = plt.subplots(figsize = (8, 6))

(pd.DataFrame(cluster_errors)
 .plot(ax = ax,
       marker = 'o', 
       zorder = 3))

ax.grid(zorder = 0)
ax.legend("")
ax.set_ylim(bottom = 0)
ax.set_xlabel('Number of Clusters (K)')
ax.set_ylabel('Centroid Inertia')
ax.set_title('Elbow Plot for Daily Profiles')
```

Performance appears to approach roughly 60k asymptotically, so for the sake of interpretation, let's proceed with K = 6 clusters. 

```python
n_clusters = 6

kmeans = KMeans(n_clusters=n_clusters,
                n_init=10)
predicted_clusters = kmeans.fit_predict(scaled_readings_pca_18)

fig, ax = plt.subplots(3, 2, figsize=(12, 8))
ax = ax.flatten()

# to visualize the clusters, loop over the clusters (i.e. unique labels of predictions)
for idx in np.unique(predicted_clusters): 

    # get the indices of the elements (rows) that belong to the same cluster 
    indices = np.where(predicted_clusters == idx)[0]
    
    # use the indices to get the relevant profiles 
    true_profiles = complete_days_df.pivot(index = ["meter_identifier", "date"], columns = "time", values = "kWh_per_half_hour").iloc[indices].reset_index(drop=True)
    
    # plot all profiles of a cluster and their mean profile
    true_profiles.T.plot(ax = ax[idx], color='blue', alpha=0.2, linewidth=0.5)
    true_profiles.mean().plot(ax = ax[idx], color='black', linewidth=2)
    
    # adjust additional formatting
    ax[idx].legend("")
    ax[idx].set_ylabel('Energy (kWh)')
    ax[idx].set_xlabel("")
    # ax[idx].set_xticks(ax[idx].get_xticks())
    ax[idx].set_xticks([time.strftime("%H:%M") for time in true_profiles.T.index][::6])    
    ax[idx].set_title('Cluster {} - Observations: {}'.format(idx, len(indices)))
    # ax[idx].set_xlim(0, 96)

plt.tight_layout()
```

With the mean profile for each cluster plotted on a more visible scale:

```python
fig, ax = plt.subplots(3, 2, figsize=(12, 8))
ax = ax.flatten()

for idx in np.unique(predicted_clusters): 

    # get the indices of the elements (rows) that belong to the same cluster 
    indices = np.where(predicted_clusters == idx)[0]
    
    # use the indices to get the relevant profiles 
    true_profiles = complete_days_df.pivot(index = ["meter_identifier", "date"], columns = "time", values = "kWh_per_half_hour").iloc[indices].reset_index(drop=True)
    
    # plot all profiles of a cluster and their mean profile
    true_profiles.mean().plot(ax = ax[idx], color='black', linewidth=2)
    
    # adjust additional formatting
    ax[idx].legend("")
    ax[idx].set_ylabel('Energy (kWh)')
    ax[idx].set_xlabel("")
    # ax[idx].set_xticks(ax[idx].get_xticks())
    ax[idx].set_xticks([time.strftime("%H:%M") for time in true_profiles.T.index][::6])    
    ax[idx].set_title('Cluster {} - Observations: {}'.format(idx, len(indices)))
    ax[idx].grid()
    # ax[idx].set_xlim(0, 96)

plt.tight_layout()
```

Although there remains a lot of intra-group variability, there is a strong case that cluster 0 include those who charge EVs overnight, based on the disproportionally high consumption rates early in the day. Unfortunately this dataset does not contain any metadata on each meter, so we're unable to say with certainty how the load may be disaggregated by device. 

```python
def plot_silhouette_values(pivoted_readings, predicted_clusters, n_clusters):
    """
    Adapted from the sklearn docs.
    """
    
    sil_sample_values = silhouette_samples(pivoted_readings,
                                           predicted_clusters)
    
    fig, ax = plt.subplots(figsize = (10, 8))
    
    y_lower = 10
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        ith_cluster_silhouette_values = sil_sample_values[predicted_clusters == i]
    
        ith_cluster_silhouette_values.sort()
    
        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i
    
        color = cm.nipy_spectral(float(i) / n_clusters)
        ax.fill_betweenx(
            np.arange(y_lower, y_upper),
            0,
            ith_cluster_silhouette_values,
            facecolor=color,
            edgecolor=color,
            alpha=0.7,
        )
    
        # Label the silhouette plots with their cluster numbers at the middle
        ax.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
    
        # Compute the new y_lower for next plot
        y_lower = y_upper + 10
    
    ax.set_title("Silhouette Scores Across Clusters")
    ax.set_xlabel("Silhouette Coefficient Values")
    ax.set_ylabel("Cluster label")
    ax.set_yticks([])
    ax.grid()

plot_silhouette_values(pivoted_readings = complete_days_df.pivot(index = ["meter_identifier", "date"], columns = "time", values = "kWh_per_half_hour").reset_index(drop=True),
                      predicted_clusters = predicted_clusters,
                      n_clusters = n_clusters)
```

Clutering across days is not very effective! Other than for cluster 1, all clusters suffer from very low scores, indicating pretty ineffective classification.

Let us try to classify each meter. Although crude, we'll start by taking the mean of each period's readings over all of the complete days available per meter to give a unified sample to cluster against.

6 clusters still looks like the inflection point.

## Octopus Agile Tariff

We will now retrieve the Octopus Agile half-hourly tariff rates for comparison against the cheapest fixed and variable tariffs available at that time, and see which meter clusters could benefit the most from switching to agile. 

```python
agile_rates = httpx.get(os.environ["OCTOPUS_TARIFF_URL"] + "/standard-unit-rates/", params={"period_from": "2017-11-05", "period_to": "2017-11-23", "page_size": 1200})
agile_rates.json()["results"][:3]
```

```python
agile_rates_df = pd.DataFrame.from_dict({pd.to_datetime(period["valid_from"]): period["value_inc_vat"] for period in agile_rates.json()["results"]}, 
                                        orient = "index",
                                        columns = ["unit_rate"])
agile_rates_df.head()
```

```python
# get the standing charge also

agile_standing_charge = httpx.get(os.environ["OCTOPUS_TARIFF_URL"] + "/standing-charges/", params={"period_from": "2017-11-05", "period_to": "2017-11-23", "page_size": 1200})
agile_standing_charge.json()["results"][:3]
```

Define a function to calculate the cost of a given day of meter readings under Octopus Agile. 

The average variable tariff on the market in 2017 gave a unit rate and standing charge of 24.6p/kWh and 33.4 p/day respectively, so we will use this as the baseline cost against which we can compare the Octopus Agile tariff.

```python
benchmark_unit_rate = 24.6
benchmark_standing_charge = 33.4

agile_standing_charge = 21.0
```

```python
def calculate_agile_cost(meter_readings, 
                         agile_rates_df,
                         benchmark_unit_rate = benchmark_unit_rate,
                         benchmark_standing_charge = benchmark_standing_charge,
                         agile_standing_charge = agile_standing_charge,
                         plot = False,
                         custom_fig_size = (8, 6)):
    """
    Joins the agile rates dataframe against interval_start 
    to create a cost profile for the given meter readings.
    The Octopus Agile standing charge was 21p/day, added on
    a per-period basis.
    """

    cost_profile = meter_readings.set_index("interval_start").join(agile_rates_df)
    cost_profile["agile_cost"] = cost_profile.kWh_per_half_hour * cost_profile.unit_rate + agile_standing_charge/48
    cost_profile["benchmark_svt_cost"] = cost_profile.kWh_per_half_hour * benchmark_unit_rate + benchmark_standing_charge/48

    if plot:

        fig, ax = plt.subplots(figsize = custom_fig_size)

        cost_profile.agile_cost.plot(
            ax = ax,
            zorder = 3
        )

        cost_profile.benchmark_svt_cost.plot(
            ax = ax,
            zorder = 3,
            color = "r",
            alpha = 0.8
        )

        ax.grid(zorder = 0)
        ax.set_ylabel("Cost (p)")
        ax.set_xlabel("")
        ax.legend(bbox_to_anchor=(1, 1), 
                  loc='upper left',
                  labels = ["Agile Cost", "Benchmark Cost"])
        
        plt.show()
        
    return cost_profile
```

```python
test_cost_profile = calculate_agile_cost(complete_days_df.iloc[:48], agile_rates_df, plot = True)
```

We will calculate the percentage saving for each customer, and see if that aligns with membership of particular clusters:

```python
cost_profile = calculate_agile_cost(complete_days_df, agile_rates_df)

cost_profile_grouped = cost_profile.groupby("meter_identifier")[["agile_cost", "benchmark_svt_cost"]].sum()
cost_profile_grouped["percentage_saving"] = round(100 * (cost_profile_grouped.benchmark_svt_cost - cost_profile_grouped.agile_cost ) / cost_profile_grouped.benchmark_svt_cost, 1)
```

```python
fig, ax = plt.subplots(figsize = (8, 6))

cost_profile_grouped.percentage_saving.plot.hist(bins=25, 
                                                 ax = ax,
                                                 zorder = 3)

ax.legend("")
ax.grid(zorder = 0)
ax.set_xlabel("Percentage Cost Saving")
ax.set_ylabel("Number of Meters")
ax.set_title("Histogram of Percentage Savings per Meter against Octopus Agile")

# highlight median of means
ax.axvline(x = cost_profile_grouped.percentage_saving.median(), 
           ymin = 0, 
           ymax = 60, 
           color = "r",
           zorder = 4)

print(f"Median saving: £{(cost_profile_grouped.benchmark_svt_cost - cost_profile_grouped.agile_cost).median()/100:.2f}")
```

Every meter in the sample could benefit by using Octopus Agile! But which clusters save the most?

```python
cost_profile_grouped["cluster_id"] = predicted_clusters

def plot_savings_by_cluster(cost_profile_grouped):
    
    fig, ax = plt.subplots(figsize = (18, 8))

    (cost_profile_grouped.reset_index()
                         .sort_values(["cluster_id", "percentage_saving"])
                         .set_index(["cluster_id"])
                         [['percentage_saving']]
                         .plot.bar(ax = ax,
                                   zorder = 3))
    
    ax.legend("")
    ax.set_ylabel("Percentage Saving per Meter")
    ax.set_xlabel("Cluster ID")
    ax.set_xticks(np.arange(0, len(cost_profile_grouped), 4),
                  cost_profile_grouped.reset_index().sort_values("cluster_id").cluster_id[::4].values)
    ax.xaxis.set_tick_params(rotation=90)
    ax.grid(zorder = 0,
           axis = "y")

plot_savings_by_cluster(cost_profile_grouped)
```

Fairly tricky to tell with so few samples, but it appears the sharper end of cluster 1 might have an edge, although it also contains some of those that save the least. A wider sample size would certainly help establish any causality, although it follow that those with a disporportionate consumption profile in the early hours of the morning stand to save the most on Agile compared to a flat SVT rate, take our sole member of cluster one for example:
